Заполним их медианными и проверим, не осталось ли больше в таблице нулевых значений.

```python
##df['ceiling_height'].median()
####df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
#df['ceiling_height'].isnull().count()
```

Теперь разберёмся с выбивающимися за границы здравого смысла значения - все, что превышают 20.

```python
df.loc[df['ceiling_height']>20,'ceiling_height'] =  df['ceiling_height']/10
drop_index = df[df['ceiling_height']>4].index
df.drop(drop_index,inplace = True)
ceil_median = df['ceiling_height'].median()
df['ceiling_height'] =  df['ceiling_height'].fillna(value = ceil_median)
#print(df['ceiling_height'])
print(df)
```

Заполним недостающие значения в столбце floors_total значениями со столбца floor+1, чтобы они таким образом попали в категорию "другой этаж", рассматриваемую ниже. Сразу проверим, сколько теперь не-нулевых значений в столбце.

```python
floor = df['floor']+1
df['floors_total'] =  df['floors_total'].fillna(value = floor)
df['floors_total'].count()
```

Площадь кухни отсутствует для квартир-студий и квартир с открытой планировкой. Заполним эти пропуски нулями.
Для квартир обычного типа площадь кухни заполним, рассчитав её пропорционально площади жилья. Для этого найдём по таблице отношение кухни к общей площади и возьмём медиану от этих значений.

```python
#df[pd.isnull(df['kitchen_area'])]
df.loc[(df['open_plan']==True) | (df['studio'] ==True),'kitchen_area']=0
#df['kitchen_to_total_area']=df['kitchen_area']/df['total_area']
#kitchen_to_total_area_median = df['kitchen_to_total_area'].median()
#df['kitchen_area']=df['kitchen_area'].fillna(value = df['total_area']*kitchen_to_total_area_median 
#df['kitchen_to_total_area']=df['kitchen_area']/df['total_area']
```

```python
df.info()
```

Посмотрим, много ли совпадающих отсутствующих значений в этих столбцах.

Заполним отсутствующие значения для совпадающих по этим двум столбцам ячеек медианой - для простоты дальнейших подсчётов.

```python
#df['kitchen_area'] = df['kitchen_area'].apply(lambda x: x.fillna(value=0,inplace = True))
median_kitchen = df['kitchen_area'].median()
median_living =  df['living_area'].median()
df['kitchen_area'] = df['kitchen_area'].fillna(median_kitchen)
df['living_area'] = df['living_area'].fillna(median_living)
#df[(df['kitchen_area'].isnull()==True) & (df['living_area'].isnull()==True)].info()
#df['living_area']=df['living_area'].astype(int)
#df['kitchen_area']=df['kitcehn_area'].astype(int)
#df['total_area']=df['total_area'].astype(int)

```

Разберёмся теперь со столбцом балконов. Если количество балконов не указано, можно считать, что их нет. Соответственно, заполним все пропуски нулями.

```python
df['balcony'] = df['balcony'].fillna(0)
df.info()
```

Далее заполним пропуски в столбце is_apartment, предполагая, что если не указано, что аппартаменты, мы считаем, что их нет.

```python
df['is_apartment']=df['is_apartment'].fillna(value = False)
```

Так как данные по аэропортам, расстоянию до центра Санкт-Петербурга, количеству парков и водоёмов отсутствует для удалённых от Санкт-Петербурга места, заполним их значением -1, чтобы не путать с нулевым значением (нет парков, водоёмов)

```python
df['airports_nearest']=df['airports_nearest'].fillna(value = -1)
df['cityCenters_nearest']=df['cityCenters_nearest'].fillna(value = -1)
df['parks_around3000']=df['parks_around3000'].fillna(value = -1)
df['parks_nearest']=df['parks_nearest'].fillna(value = -1)
df['ponds_around3000'] = df['ponds_around3000'].fillna(value = -1)
df['ponds_nearest']=df['ponds_nearest'].fillna(value = -1)
```

В столбце days_exposition 13%  значений отсутствует - возможно потому, что квартира на момент выгрузки базы данных не была ещё продана. Эти строки можно удалить, так как это не повлияет на показатели рыночной стоимости.

```python
df=df.dropna(subset =['days_exposition'])
```

Осталось разобраться со столбцом locality_name. Населённый пункт можно вычислить, исходя из расстояния от центра города. Из сводных данных видно, что самый ближайший пригород находится на расстоянии 18700 метров, значит всё, что ближе, считаем, что находится в Санкт-Петербурге. 

```python
df.query('cityCenters_nearest>0').pivot_table(index = 'locality_name',values = 'cityCenters_nearest',aggfunc = 'max').reset_index().sort_values(by = 'cityCenters_nearest',ascending = True).head(1)
df['locality_name']=df['locality_name'].fillna(value = 'нет данных')
df.loc[(df['cityCenters_nearest']<=18700)&(df['locality_name']=='нет данных'),'locality_name']='Санкт-Петербург'
df.info()
```

## Расчёты и добавление результатов в таблицу

Добавим новый столбец square_meter_price

```python
df['square_meter_price']= df['last_price']/df['total_area']
print(df)
```

Создадим ещё три столбца - срез столбца с датой по году, месяцу и дню недели.

```python
df['first_day_exposition']=pd.to_datetime(df['first_day_exposition'],format = '%Y-%m-%dT%H:%M:%S')
df['first_day_exposition_year'] = df['first_day_exposition'].dt.year
df['first_day_exposition_month'] = df['first_day_exposition'].dt.month
df['first_day_exposition_weekday'] = df['first_day_exposition'].dt.weekday
print(df['first_day_exposition_year'])
print(df['first_day_exposition_month'])
print(df['first_day_exposition_weekday'])
```

Напишем функцию, определяющую, на первом, последнем или другом этаже находится квартира и добавляем столбец к столбцу этажей.

```python
def floor_index (row):
    if row['floor'] == 1:
        return 'Первый этаж'
    elif row['floor'] == row['floors_total']:
        return 'Последний этаж'
    return 'Другой этаж'
df['floor_index_new']=df.apply(floor_index,axis = 1)
print(df['floor_index_new'])
```

Добавим два столбца, показывающие соотношение жилой площади к общей и площади кухни к общей.

```python
df['living_total_ratio']=df['living_area']/df['total_area']
df['kitchen_total_ratio']=df['kitchen_area']/df['total_area']
```

Построим таблицу коэффициентов корреляции, показывающую зависимость от стоимости квартиры от основных параметров.

```python
corr_rooms_centre_meter = df4_spb_centre['sq_meter_price'].corr(df4_spb_centre['rooms'])
corr_rooms_meter = df4_spb['sq_meter_price'].corr(df4_spb['rooms'])
corr_rooms_all_meter = df4['sq_meter_price'].corr(df4['rooms'])

corr_rooms_centre = df4_spb_centre['last_price'].corr(df4_spb_centre['rooms'])
corr_rooms= df4_spb['last_price'].corr(df4_spb['rooms'])
corr_rooms_all = df4['last_price'].corr(df4['rooms'])

corr_total_area_centre_meter = df4_spb_centre['sq_meter_price'].corr(df4_spb_centre['total_area'])
corr_total_area_meter= df4_spb['sq_meter_price'].corr(df4_spb['total_area'])
corr_total_area_all_meter = df4['sq_meter_price'].corr(df4['total_area'])

corr_total_area_centre = df4_spb_centre['last_price'].corr(df4_spb_centre['total_area'])
corr_total_area= df4_spb['last_price'].corr(df4_spb['total_area'])
corr_total_area_all = df4['last_price'].corr(df4['total_area'])

corr_km_centre = df4_spb_centre['last_price'].corr(df4_spb_centre['cityCenters_km'])
corr_km= df4_spb['last_price'].corr(df4_spb['cityCenters_km'])
corr_km_all = df4['last_price'].corr(df4['cityCenters_km'])

corr_month_centre  = df4_spb_centre['last_price'].corr(df4_spb_centre['first_day_exposition_month'])
corr_month= df4_spb['last_price'].corr(df4_spb['first_day_exposition_month'])
corr_month_all = df4['last_price'].corr(df4['first_day_exposition_month'])

corr_floor_centre  = df4_spb_centre['last_price'].corr(df4_spb_centre['floor_index_new'].astype('category').cat.codes)
corr_floor= df4_spb['last_price'].corr(df4_spb['floor_index_new'].astype('category').cat.codes)
corr_floor_all = df4['last_price'].corr(df4['floor_index_new'].astype('category').cat.codes)

corr_ceil_centre  = df4_spb_centre['last_price'].corr(df4_spb_centre['ceiling_height'])
corr_ceil= df4_spb['last_price'].corr(df4_spb['ceiling_height'])
corr_ceil_all = df4['last_price'].corr(df4['ceiling_height'])

data = [['Количество комнат', corr_rooms_centre, corr_rooms, corr_rooms_all],
        ['Количество комнат(цена кв.м.)',corr_rooms_centre_meter,corr_rooms_meter,corr_rooms_all_meter],
        ['Площадь квартиры',corr_total_area_centre,corr_total_area,corr_total_area_all],
        ['Площадь квартиры (цена кв.м.)',corr_km_centre,corr_km,corr_km_all],
        ['Месяц размещения',corr_month_centre,corr_month,corr_month_all],
        ['Высота потолков',corr_ceil_centre,corr_ceil,corr_ceil_all],
        ['Этаж',corr_floor_centre,corr_floor,corr_floor_all],
       ]
columns = ['Зависимость цены от', 'Центр до 10 км', 'Санкт-Петербург','Область']
result = pd.DataFrame(data = data, columns = columns)
result
```
